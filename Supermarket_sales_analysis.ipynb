# Data Handling:

```python
#First we handle Unit Price by using Mean Value:
avg_unit_price = Data["Unit price"].astype("float").mean(axis=0)
Data["Unit price"].replace(np.nan, avg_unit_price, inplace=True)
```

```python
#Now for quantity we can use mode:
from scipy import stats
mode=stats.mode(Data['Quantity'])

print(mode)
```

```python
Data['Quantity'].replace(np.nan, "mode", inplace=True)
```

```python
#We can drop the remaining values:
Data.dropna(inplace=True)
```

```python
#Finally we can see No null values:
Data.isnull().sum()
```

```python
#Let's check the correlation of the data:
Data.corr()
```

```python
np.round(Data.corr(),2)
```

# Exploratory Data Analysis(EDA): 

```python
plt.figure(dpi=100)
sns.heatmap(np.round(Data.corr(),2),annot=True)
plt.show()
```

```python
#Gross Income vs. Tax 5%:
plt.figure(figsize=(10,5),dpi=100)
sns.regplot(x='Tax 5%',y='gross income',data=Data,color='Red')
plt.xlabel('Tax 5%')
plt.ylabel('Gross Income')
plt.title('Tax 5% v Gross Income',fontsize=14)
plt.show()
```

```python
#Quantity vs. Cost of Goods Sales:
plt.figure(figsize=(10,5),dpi=80)
sns.regplot(x='Quantity',y='cogs',data=Data,color='blue')
plt.xlabel('Quantity')
plt.ylabel('Cost of Goods Sale')
plt.title('Quantity vs. Cost of Goods Sale',fontsize=14)
plt.show()
```

```python
#Unit Price vs. Gross Income:
plt.figure(figsize=(10,5),dpi=80)
sns.regplot(x='Unit price',y='gross income',data=Data,color='green')
plt.xlabel('Unit Price')
plt.ylabel('Gross Income')
plt.title('Unit Price v Gross Income',fontsize=14)
plt.show()
```

What does the customer rating looks like?

```python
#Customer Rating and its skewness wrt quartiles:
sns.displot(Data['Rating'])
plt.axvline(x=np.mean(Data['Rating']), c='red', ls='--', label='mean')
plt.axvline(x=np.percentile(Data['Rating'],25),c='black', ls='--', label = '25th percentile:Q1')
plt.axvline(x=np.percentile(Data['Rating'],75),c='cyan', ls='--',label = '75th percentile:Q3' )
plt.legend()
```
